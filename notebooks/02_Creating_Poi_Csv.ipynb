The following code improves over the previous code by reducing duplicate API calls and gets the viewport cordinates instead of calculating radius.  It first creates a Google Maps client object and then iterates over the zip codes and calls the places() method on the client object for each zip code. The places() method returns a list of POIs for the given zip code. It then adds each POI to a list and saves the list to a CSV file.

The code also checks the status of the Places API response before adding the POIs to the list. This will prevent the code from adding POIs that are not found. It also adds a short delay between API calls which will help to prevent the code from exceeding the API's rate limit.



```python


class POIGenerator(object):
    TYPES = [
    "art_gallery",
    "atm",
    "bakery",
    "bar",
    "beauty_salon",
    "bicycle_store",
    "book_store",
    "cafe",
    "car_rental",
    "clothing_store",
    "convenience_store",
    "gym",
    "hair_care",
    "home_goods_store",
    "library",
    "liquor_store",
    "laundry",
    "movie_theater",
    "night_club",
    "park",
    "parking",
    "pet_store",
    "restaurant",
    "real_estate_agency",
    "spa",
    "store",
    "storage",
    "supermarket",
    "university",
    "train_station",
    "subway_station",
    "transit_station",
    "bus_station",
    "light_rail_station",
    ]
    
    USEFUL_FIELDS = ("place_id", "name", "rating", "user_ratings_total")
    EXTRA_FIELDS = ("latitude", "longitude", "poi_type", "zip_code", "southwest_lat", "southwest_lng", "northeast_lat", "northeast_lng")


    def __init__(self, key, poi_type, zip_codes, pages_to_go=5):
        """
        Args:
            key: Google Maps Key
            zip_codes: List of zip codes to search POIs
            pages_to_go: Number of pages to get from results
        """
        self.key = key
        self.pages_to_go = pages_to_go
        self.poi_list = []
        self.poi_type = poi_type
        self.zip_codes = zip_codes

    def __enter__(self):
        self.gmaps = googlemaps.Client(key=self.key)
        return self

    def __exit__(self, exc_type, exc_val, exc_tb):
        if self.gmaps:
            self.gmaps.session.close()
                    

    def add_poi(self, results, zip_code):
        for r in results:
            new_r = r.copy()
            new_r["latitude"] = r["geometry"]["location"]["lat"]
            new_r["longitude"] = r["geometry"]["location"]["lng"]
            new_r["poi_type"] = self.poi_type
            new_r["zip_code"] = zip_code

            if "viewport" in r["geometry"]:
                viewport = r["geometry"]["viewport"]
                new_r["southwest_lat"] = viewport["southwest"]["lat"]
                new_r["southwest_lng"] = viewport["southwest"]["lng"]
                new_r["northeast_lat"] = viewport["northeast"]["lat"]
                new_r["northeast_lng"] = viewport["northeast"]["lng"]
            else:
                new_r["southwest_lat"] = None
                new_r["southwest_lng"] = None
                new_r["northeast_lat"] = None
                new_r["northeast_lng"] = None

            self.poi_list.append(new_r)

    def fetch_pois(self, zip_code):
        current_page = 0
        next_page_token = None

        while current_page < self.pages_to_go:
            params = {"query": f"{self.poi_type} in {zip_code}"}
            if next_page_token:
                params["page_token"] = next_page_token

            places = self.gmaps.places(**params)
            if 'status' in places and places['status'] == 'OK':
                self.add_poi(places["results"], zip_code)

                next_page_token = places.get("next_page_token", None)
                if not next_page_token:
                    break
            else:
                print(f"Error fetching POIs for zip code {zip_code}: {places}")
                break

            time.sleep(0.5)  # Short delay between API calls
            current_page += 1

    def get_poi_list(self):
        return self.poi_list
    
    def save_to_csv(self, filename):
        if not self.poi_list:
            return

        fieldnames = set(self.poi_list[0].keys())
        with open(filename, "w") as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(self.poi_list)



# zip_codes = ["30021"]  #  zip codes here

for poi_type in POIGenerator.TYPES:
    with POIGenerator(
        key="<removed>",
        poi_type=poi_type,
        zip_codes=zip_codes,
    ) as poi_generator:
        for zip_code in zip_codes:
            poi_generator.fetch_pois(zip_code)
        
        poi_list = poi_generator.get_poi_list()
        poi_generator.save_to_csv(f"poi_30002_30038_{poi_type}.csv")
```
